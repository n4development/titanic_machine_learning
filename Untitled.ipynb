## Feature Engineering and EDA [ Categorical , numerical , Null ]

```python
def substrings_in_string(big_string, substrings):
    for substring in substrings:
        if big_string.find(substring) != -1:
            return substring
    print(big_string)
    return np.nan

title_list=['Mrs', 'Mr', 'Master', 'Miss', 'Major', 'Rev',
                    'Dr', 'Ms', 'Mlle','Col', 'Capt', 'Mme', 'Countess',
                    'Don', 'Jonkheer']

df['Title']=df['Name'].map(lambda x: substrings_in_string(x, title_list))
 
#replacing all titles with mr, mrs, miss, master
def replace_titles(x):
    title=x['Title']
    if title in ['Don', 'Major', 'Capt', 'Jonkheer', 'Rev', 'Col']:
        return 'Mr'
    elif title in ['Countess', 'Mme']:
        return 'Mrs'
    elif title in ['Mlle', 'Ms']:
        return 'Miss'
    elif title =='Dr':
        if x['Sex']=='male':
            return 'Mr'
        else:
            return 'Mrs'
    else:
        return title
df['Title']=df.apply(replace_titles, axis=1)

#Creating new family_size column
df['Family_Size']=df['SibSp']+df['Parch']
df['Age*Class']=df['Age']*df['Pclass']
df['Fare_Per_Person']=df['Fare']/(df['Family_Size']+1)
#df.drop('PassengerId', axis=1, inplace=1) 
df['Sex'].replace('male',1,inplace=True)
df['Sex'].replace('female',2,inplace=True)
average_age   = df["Age"].mean()
std_age       = df["Age"].std()
count_nan_age = df["Age"].isnull().sum()
# formula rand( avg - std , avg + std ) 
random_age = np.random.randint(average_age - std_age, average_age + std_age, size = count_nan_age)
df['Age'].dropna()
df["Age"][np.isnan(df["Age"])] = random_age
df['Age'] = df['Age'].astype(int)

scatter_matrix(df[["Survived","Pclass","Sex","Fare_Per_Person","Family_Size"]],figsize=[30,30],marker='x',diagonal='kde',c=df.Survived.apply(lambda x:colors[x]))
df.info()
```

```python
%%time 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.cross_validation import KFold
from sklearn.preprocessing import scale

imputed_features_training = df[['Family_Size', 'Age*Class', 'Fare_Per_Person','Sex']]
X_train = imputed_features_training.values
y_train = df['Survived'].values

folds = KFold(y_train.shape[0], n_folds=50, shuffle=True)
cv_accuracies = []
for trining_idx, testing_idx in folds:
    X_train_cv = X_train[trining_idx]
    y_train_cv = y_train[trining_idx]
    
    X_test_cv = X_train[testing_idx]
    y_test_cv = y_train[testing_idx]
    
    logistic_regression = LogisticRegression()
    logistic_regression.fit(scale(X_train_cv), y_train_cv)
    y_predict_cv = logistic_regression.predict(scale(X_test_cv))
    current_accuracy = accuracy_score(y_test_cv, y_predict_cv)
    cv_accuracies.append(current_accuracy)
    print('cross validation accuracy:',current_accuracy)
    
print('---------------------------------------')
print('average corss validation accuracy',sum(cv_accuracies)/len(cv_accuracies))    
#print '---------------------------------------'
```

```python
def handle_missing_data_Title_for_Age(df, keep_binary=False, keep_scaled=False):
    
    df['Names'] = df['Name'].map(lambda x: len(re.split(' ', x)))
    
    # person title
    df['Title'] = df['Name'].map(lambda x: re.compile(", (.*?)\.").findall(x)[0])
    
    # group low-occuring, related titles together
    df['Title'][df.Title == 'Jonkheer'] = 'Master'
    df['Title'][df.Title.isin(['Ms','Mlle'])] = 'Miss'
    df['Title'][df.Title == 'Mme'] = 'Mrs'
    df['Title'][df.Title.isin(['Capt', 'Don', 'Major', 'Col', 'Sir'])] = 'Sir'
    df['Title'][df.Title.isin(['Dona', 'Lady', 'the Countess'])] = 'Lady'
    
    # Build binary features
    if keep_binary:
        df = pd.concat([df, pd.get_dummies(df['Title']).rename(columns=lambda x: 'Title_' + str(x))], axis=1)
    
    # process scaling
    if keep_scaled:
        scaler = preprocessing.StandardScaler()
        df['Names_scaled'] = scaler.fit_transform(df['Names'])
    
    if keep_binary:
        df['Title_id'] = pd.factorize(df['Title'])[0]+1
    
    if keep_binary and keep_scaled:
        scaler = preprocessing.StandardScaler()
        df['Title_id_scaled'] = scaler.fit_transform(df['Title_id'])
    return df
```

```python
def handle_missing_data_Cabin(df, keep_binary=False, keep_scaled=False):   
    # Replace missing values with "NA"
    df['Cabin'][df.Cabin.isnull()] = 'NA'
    
    # create feature for the alphabetical part of the cabin number
    df['CabinLetter'] = df['Cabin'].map( lambda x : cabin_letter(x))
    df['CabinLetter'] = pd.factorize(df['CabinLetter'])[0]
    
    # create binary features for each cabin letters
    if keep_binary:
        cletters = pd.get_dummies(df['CabinLetter']).rename(columns=lambda x: 'CabinLetter_' + str(x))
        df = pd.concat([df, cletters], axis=1)
    
    # create feature for the numerical part of the cabin number
    df['CabinNumber'] = df['Cabin'].map( lambda x : cabin_number(x)).astype(int) + 1
    # scale the number to process as a continuous feature
    if keep_scaled:
        scaler = preprocessing.StandardScaler()
        df['CabinNumber_scaled'] = scaler.fit_transform(df['CabinNumber'])
    return df;


def cabin_letter(cabin):
    """
    Find the letter component of the Cabin variable
    """
    match = re.compile("([a-zA-Z]+)").search(cabin)
    if match:
        return match.group()
    else:
        return 'U'


def cabin_number(cabin):
    """
    Find the number component of the Cabin variable
    """
    match = re.compile("([0-9]+)").search(cabin)
    if match:
        return match.group()
    else:
        return 0
```
